Question 1: What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.
Answer:

Artificial Intelligence (AI) is a broad field of computer science that focuses on creating systems capable of performing tasks that normally require human intelligence, such as decision-making, problem-solving, speech recognition, and image understanding. AI aims to make machines think and act intelligently.

Machine Learning (ML) is a subset of AI that enables computers to learn from data without being explicitly programmed. ML algorithms identify patterns in data and use those patterns to make predictions or decisions. Examples include spam detection, recommendation systems, and fraud detection.

Deep Learning (DL) is a specialized branch of Machine Learning that uses artificial neural networks with multiple layers to learn complex patterns from large amounts of data. It is widely used in image recognition, natural language processing, and autonomous vehicles.

Data Science is an interdisciplinary field that involves collecting, cleaning, analyzing, and interpreting data to extract meaningful insights. It combines statistics, programming, machine learning, and domain knowledge to solve business and research problems.

In summary, AI is the broadest field, ML is a subset of AI, DL is a subset of ML, and Data Science focuses on extracting knowledge and insights from data using various analytical techniques.



Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?
Answer:

Overfitting occurs when a machine learning model learns the training data too well, including noise and random fluctuations. As a result, the model performs very well on training data but poorly on new or unseen data. Overfitting usually happens when a model is too complex for the amount of available data.

Underfitting occurs when a model is too simple to capture the underlying patterns in the data. In this case, the model performs poorly on both training and testing datasets.

These problems can be detected by comparing training and testing performance. If training accuracy is very high but testing accuracy is low, the model is likely overfitting. If both training and testing accuracies are low, the model is likely underfitting.

Several techniques can help prevent these issues:

1.Cross-validation helps evaluate model performance on different subsets of data.

2.Regularization techniques such as L1 and L2 reduce model complexity.

3.Feature selection removes unnecessary features.

4.Increasing training data helps reduce overfitting.

5.Choosing an appropriate model complexity helps balance bias and variance.

A good machine learning model achieves a balance between overfitting and underfitting and generalizes well to unseen data.

Question 3: How would you handle missing values in a dataset? Explain at least three methods with examples.
Answer:

Missing values are common in real-world datasets and must be handled carefully before building machine learning models.

1. Deletion Method

Rows or columns containing missing values can be removed when the number of missing entries is small.

Example:df.dropna(inplace=True)

This method is simple but may result in loss of valuable information if many records are removed.

2. Mean or Median Imputation

For numerical data, missing values can be replaced with the mean or median of the column.

Example:

df['Age'].fillna(df['Age'].mean(), inplace=True)

Median is preferred when the data contains outliers.

3. Predictive Modeling

Machine learning algorithms can predict missing values using other available features.

Example:

A regression model can estimate missing salary values based on age, education, and experience.

This method often produces more accurate estimates but requires additional computation.

Using the appropriate method depends on the amount, type, and pattern of missing data in the dataset.

Question 4: What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).
Answer:

An imbalanced dataset is a dataset where one class contains significantly more observations than another class. This is common in fraud detection, disease diagnosis, and spam detection problems.

For example, in a dataset of 10,000 transactions, if only 100 are fraudulent and 9,900 are legitimate, the dataset is highly imbalanced.

1. Random Oversampling

Random oversampling increases the number of minority-class samples by duplicating existing observations.

Example:from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler()
X_resampled, y_resampled = ros.fit_resample(X, y)

2. SMOTE (Synthetic Minority Oversampling Technique)

SMOTE creates synthetic samples instead of duplicating existing records.

Example:

from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

SMOTE generally produces better results because it generates new, realistic examples for the minority class.

These techniques help machine learning models learn equally from both classes and improve classification performance.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Answer:

Label Encoding and One-Hot Encoding are techniques used to convert categorical data into numerical form so that machine learning algorithms can process it.

Label Encoding assigns a unique numerical value to each category.

Example:

Color	Encoded Value
Red	0
Blue	1
Green	2

This method is memory-efficient and works well when categories have a natural order, such as:

Low = 1
Medium = 2
High = 3

However, for unordered categories, Label Encoding may incorrectly imply a relationship between categories.

One-Hot Encoding creates a separate binary column for each category.

Example:

Color	Red	Blue	Green
Red	1	0	0
Blue	0	1	0
Green	0	0	1

One-Hot Encoding removes any artificial ordering between categories and is generally preferred for nominal variables such as city names, product categories, or colors.

Therefore, Label Encoding should be used for ordinal categorical data, while One-Hot Encoding is more suitable for nominal categorical data where no natural order exists.

Question 7: Google Play Store Dataset

```python
import pandas as pd

# Load dataset
df = pd.read_csv("googleplaystore.csv")

# Convert Rating to numeric
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")

# Remove invalid category
df = df[df["Category"] != "1.9"]

# Calculate average rating by category
avg_rating = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)

print("Top 5 Categories by Average Rating:")
print(avg_rating.head())

print("\nBottom 5 Categories by Average Rating:")
print(avg_rating.tail())
```

Analysis

The analysis shows that EVENTS, EDUCATION, and ART_AND_DESIGN applications have the highest average ratings among all categories. These apps are generally focused on providing specific services or educational content, which leads to higher user satisfaction and positive reviews.

In contrast, DATING, TOOLS, and MAPS_AND_NAVIGATION categories have the lowest average ratings. Dating applications often receive complaints related to fake profiles, privacy issues, and paid features. Utility and navigation apps are used by a large number of users and are more likely to receive negative feedback regarding bugs, advertisements, accuracy, or performance problems.

Conclusion

There is a noticeable relationship between app category and user ratings. Categories such as Events, Education, and Art & Design tend to achieve higher ratings because they effectively meet user expectations. On the other hand, categories like Dating, Tools, and Maps & Navigation generally receive lower ratings due to greater user expectations, technical issues, and service-related challenges. Therefore, app category plays an important role in influencing user satisfaction and overall ratings.

Question 8: Titanic Dataset

```python
import pandas as pd

# Load dataset
df = pd.read_csv("titanic.csv")

# Survival rate by passenger class
survival_rate = df.groupby("Pclass")["Survived"].mean()

print(survival_rate)
```

Analysis

The results show that 1st Class passengers had the highest survival rate (62.96%), followed by 2nd Class passengers (47.28%), while 3rd Class passengers had the lowest survival rate (24.24%).

A possible reason is that first-class passengers were given priority during evacuation and had easier access to lifeboats because their cabins were located closer to the upper decks. Third-class passengers faced difficulties reaching lifeboats due to their lower-deck locations and overcrowding.

Conclusion

Passenger class had a significant impact on survival. The higher the class, the greater the chance of survival, with 1st Class passengers being the most likely to survive.
