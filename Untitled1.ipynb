1.What are data structures, and why are they important?

Answ. Data structures are ways of organizing and storing data so it can be accessed and modified efficiently. They are important because they improve performance, memory usage, and code organization.

2.Explain the difference between mutable and immutable data types with examples.

Mutable: Can be changed after creation (e.g., list, dictionary, set).
Immutable: Cannot be changed after creation (e.g., string, tuple).

4. Describe how dictionaries store data.

Dictionaries store data as key-value pairs. Each key is unique and maps to a value.


5. Why might you use a set instead of a list in Python?

Sets automatically remove duplicates and provide faster membership testing.

6. What is a string in Python, and how is it different from a list?

A string is a sequence of characters and is immutable, while a list can store different data types and is mutable.

7. How do tuples ensure data integrity in Python?

Since tuples are immutable, their contents cannot be accidentally modified after creation.

8. What is a hash table, and how does it relate to dictionaries in Python?

A hash table is a data structure that stores key-value pairs using hash functions. Python dictionaries are implemented using hash tables.

9. Can lists contain different data types in Python?

Yes.

my_list = [1, "Hello", 3.14, True]

10. Explain why strings are immutable in Python.

Strings are immutable to improve performance, security, and memory efficiency.

11. What advantages do dictionaries offer over lists for certain tasks?

Dictionaries provide fast lookup using keys and are suitable for storing related data.

12. Describe a scenario where using a tuple would be preferable over a list.

Storing fixed information such as coordinates:

point = (10, 20)

13. How do sets handle duplicate values in Python?

Sets automatically remove duplicate values.

{1, 2, 2, 3}  # {1, 2, 3}

14. How does the “in” keyword work differently for lists and dictionaries?

In lists, it checks values.
In dictionaries, it checks keys.

15. Can you modify the elements of a tuple? Explain why or why not.

No. Tuples are immutable, so their elements cannot be modified.

16. What is a nested dictionary, and give an example of its use case?

A dictionary inside another dictionary.

```python
student = {
    "name": "John",
    "marks": {"Math": 90, "Science": 85}
}
```

17. Describe the time complexity of accessing elements in a dictionary.

Average-case time complexity is O(1).

18. In what situations are lists preferred over dictionaries?

When order matters and data is accessed by index rather than key.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Traditionally, dictionaries did not guarantee order. Data retrieval is done using keys, not positions.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

Lists use indexes.
Dictionaries use keys.
