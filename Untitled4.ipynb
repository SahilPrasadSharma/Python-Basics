Question 1: What is the difference between multithreading and multiprocessing?

Answer:

Multithreading and multiprocessing are techniques used to execute multiple tasks simultaneously, but they work in different ways.

Multithreading involves creating multiple threads within a single process. All threads share the same memory space and resources of the process. Because threads share memory, communication between them is faster and consumes less memory. Multithreading is mainly used for I/O-bound tasks, such as file handling, web requests, and database operations, where a program spends a lot of time waiting for external resources.

For example, a web browser can use multiple threads to load images, videos, and web pages simultaneously.

Multiprocessing, on the other hand, involves creating multiple independent processes. Each process has its own memory space and resources, which makes it more secure and stable. Since processes do not share memory directly, communication between them is slower and requires special mechanisms such as pipes or queues. Multiprocessing is best suited for CPU-bound tasks, such as scientific calculations, data analysis, and image processing, where intensive computation is required.

For example, a video editing application may use multiple processes to render different parts of a video simultaneously.

In summary, multithreading is more efficient for tasks that involve waiting for input/output operations, while multiprocessing is better for tasks that require heavy computation and full utilization of multiple CPU cores.

Question 2: What are the challenges associated with memory management in Python?

Answer:

Python automatically manages memory using reference counting and garbage collection, but some challenges still exist.

One challenge is memory leaks, where objects remain in memory because references to them still exist. Another issue is circular references, where objects refer to each other, making memory cleanup more difficult. Python may also face memory fragmentation, which reduces memory efficiency over time.

Large applications that handle huge datasets can experience high memory usage, leading to slower performance. Additionally, the garbage collection process itself can consume system resources and affect performance slightly.

Therefore, developers should write efficient code, remove unnecessary references, and properly manage resources such as files and database connections to ensure effective memory management.

Question 3: Write a Python program that logs an error message to a log file when a division by zero exception occurs.

Answer:

```python
import logging

# Configure logging
logging.basicConfig(
    filename='error.log',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

try:
    result = 10 / 0
except ZeroDivisionError:
    logging.error("Division by zero occurred.")
    print("Error logged successfully.")
```

Question 4: Write a Python program that reads from one file and writes its content to another file.

Answer:

```python
# Read from source file and write to destination file

with open("source.txt", "r") as source:
    content = source.read()

with open("destination.txt", "w") as destination:
    destination.write(content)

print("File copied successfully.")
```

Question 5: Write a program that handles both IndexError and KeyError using a try-except block.

Answer:

```python
my_list = [10, 20, 30]
my_dict = {"name": "John"}

try:
    print(my_list[5])      # Causes IndexError
    print(my_dict["age"])  # Causes KeyError
except IndexError:
    print("IndexError: List index is out of range.")
except KeyError:
    print("KeyError: Key not found in dictionary.")
```

Question 6: What are the differences between NumPy arrays and Python lists?

Answer:

NumPy arrays and Python lists are both used to store collections of data, but they differ in performance and functionality.

A Python list is a built-in data structure that can store elements of different data types, such as integers, strings, and floats. Lists are flexible and easy to use, making them suitable for general-purpose programming. However, they are slower and consume more memory when working with large amounts of numerical data.

A NumPy array is provided by the NumPy library and is designed specifically for numerical and scientific computing. It stores elements of the same data type, which allows it to perform calculations much faster and use memory more efficiently. NumPy arrays also support vectorized operations, making mathematical computations simpler and quicker.

For example, multiplying every element in a Python list usually requires a loop, whereas a NumPy array can perform the operation directly on the entire array.

Question 7: Explain the difference between apply() and map() in Pandas.

Answer:

map()
Works only on a Pandas Series (single column).
Used for element-wise transformations.
Accepts functions, dictionaries, or Series.

```python
import pandas as pd

s = pd.Series([1, 2, 3])
print(s.map(lambda x: x * 2))
```

apply()
Works on both Series and DataFrames.
Can apply functions row-wise or column-wise.
More flexible than map().
Example

```python
import pandas as pd

df = pd.DataFrame({"A": [1, 2, 3]})
print(df["A"].apply(lambda x: x * 2))
```

Question 10: Calculate the correlation matrix using Seaborn and visualize it with a heatmap.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Math': [80, 90, 70, 85, 95],
    'Science': [78, 88, 72, 84, 96],
    'English': [75, 85, 68, 80, 92]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Visualize using heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix Heatmap")
plt.show()
```

Explanation:

corr() calculates correlations between columns.
heatmap() visualizes the correlation matrix.
annot=True displays correlation values inside the cells.
