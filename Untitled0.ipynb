6. Mutable and Immutable Arguments

```python
def modify_list(lst):
    lst.append(4)

def modify_num(num):
    num += 1
    print(num)

lst = [1, 2, 3]
num = 5

modify_list(lst)
modify_num(num)

print(lst)
print(num)
```

14. Reverse a String Using While Loop

```python
text = input("Enter a string: ")
i = len(text) - 1

while i >= 0:
    print(text[i], end="")
    i -= 1
```

15. Factorial Using While Loop

```python
num = int(input("Enter a number: "))

fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial =", fact)
```
